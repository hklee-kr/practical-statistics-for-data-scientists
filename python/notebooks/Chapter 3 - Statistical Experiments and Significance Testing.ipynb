# Resampling

````python
# P.121~
session_times = pd.read_csv(WEB_PAGE_DATA_CSV)
print(session_times.head())
session_times.Time = 100 * session_times.Time
````

````python
ax = session_times.boxplot(by='Page', column='Time', figsize=(4, 4)) # Page, Time: csv 파일의 컬럼명
ax.set_xlabel('')
ax.set_ylabel('Time (in seconds)')
plt.suptitle('')

plt.tight_layout()
plt.show() # B에서의 세션 시간이 더 길다
````

````python
mean_a = session_times[session_times.Page == 'Page A'].Time.mean()
mean_b = session_times[session_times.Page == 'Page B'].Time.mean()
print(mean_b - mean_a) # 차이가 유의미한가?
````

The following code is different to the R version. idx_A and idx_B are reversed.

````python
# Permutation test example with stickiness
def perm_fun(x, nA, nB): # 순열검정
    n = nA + nB # n: 총 데이터 수
    idx_B = set(random.sample(range(n), nB)) # 세션 시간 랜덤 할당
    idx_A = set(range(n)) - idx_B # set -> 비복원추출
    return x.loc[list(idx_B)].mean() - x.loc[list(idx_A)].mean()
    
nA = session_times[session_times.Page == 'Page A'].shape[0]
nB = session_times[session_times.Page == 'Page B'].shape[0]
print(perm_fun(session_times.Time, nA, nB))
````

````python
# random.seed(1)

# 순열검정을 1000번 반복
perm_diffs = [perm_fun(session_times.Time, nA, nB) for _ in range(1000)]

# 반복 후 histogram으로 표현
fig, ax = plt.subplots(figsize=(5, 5))
ax.hist(perm_diffs, bins=11, rwidth=0.9)
ax.axvline(x = mean_b - mean_a, color='black', lw=2) # 실제 관측된 세션 시간의 차이
ax.text(50, 190, 'Observed\ndifference', bbox={'facecolor':'white'})
ax.set_xlabel('Session time differences (in seconds)')
ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show() # 확률분포 범위 내에 있으므로 통계적으로 무의미
````

````python
# convert perm_diffs to numpy array to avoid problems with some Python installations
perm_diffs = np.array(perm_diffs)
print(np.mean(perm_diffs > mean_b - mean_a)) # 무작위 순열로 구현 평균 시간이 실제 시간을 넘어갈 확률
````

# Statistical Significance and P-Values

````python
# random.seed(1)

# 가격A: 전환(200) + 비전환(23,539) = 23,739
# 가격B: 전환(182) + 비전환(22,406) = 22,588

obs_pct_diff = 100 * (200 / 23739 - 182 / 22588)
print(f'Observed difference: {obs_pct_diff:.4f}%') # 0.0368%
conversion = [0] * 45945 # 비전환
conversion.extend([1] * 382) # 전환
conversion = pd.Series(conversion)

# 순열검정을 1000번 반복
perm_diffs = [100 * perm_fun(conversion, 23739, 22588) 
              for _ in range(1000)]

# 반복 후 histogram으로 표현
fig, ax = plt.subplots(figsize=(5, 5))
ax.hist(perm_diffs, bins=11, rwidth=0.9)
ax.axvline(x=obs_pct_diff, color='black', lw=2) # 실제 관측된 전환율 차이
ax.text(0.06, 200, 'Observed\ndifference', bbox={'facecolor':'white'})
ax.set_xlabel('Conversion rate (percent)')
ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show() # 확률분포 범위 내에 있으므로 통계적으로 무의미
````

## P-Value
If `np.mean` is applied to a list of booleans, it gives the percentage of how often True was found in the list (#True / #Total).

````python
print(np.mean([diff > obs_pct_diff for diff in perm_diffs])) # 순열검정으로 얻은 전환율 차이 중 실제 관측값보다 큰 경우의 비율
````

````python
# 가설이 이항분포를 따르므로 순열검정 대신 정규근사를 통해 p값을 근사

survivors = np.array([[200, 23739 - 200], [182, 22588 - 182]])
chi2, p_value, df, _ = stats.chi2_contingency(survivors)

print(f'p-value for single sided test: {p_value / 2:.4f}')
````

# t-Tests

````python
res = stats.ttest_ind(session_times[session_times.Page == 'Page A'].Time, 
                      session_times[session_times.Page == 'Page B'].Time,
                      equal_var=False)

# If 'equal_var'=True (default), perform a standard independent 2 sample test that assumes equal population variances
# If 'equal_var'=False, perform Welch’s t-test(독립표본 T검정)

print(f'p-value for single sided test: {res.pvalue / 2:.4f}')
````

````python
tstat, pvalue, df = sm.stats.ttest_ind(
    session_times[session_times.Page == 'Page A'].Time, 
    session_times[session_times.Page == 'Page B'].Time,
    usevar='unequal', alternative='smaller') # alternative(대안가설) = ‘smaller’: difference in means smaller than value

# print('mean time of the Page A', np.mean(session_times[session_times.Page == 'Page A'].Time))
# print('mean time of the Page B', np.mean(session_times[session_times.Page == 'Page B'].Time))

print('test statistic(검정통계량): ', tstat) # t-value: 두 표본 그룹 평균의 차이를 표준오차로 나눈 것
print('degrees of freedom: ', df)
print(f'p-value: {pvalue:.4f}') # 0.05보다 크므로 통계적으로 유의미하지 않다 -> 우연히 발생 가능하다
````

# ANOVA(Analysis of Variance)

````python
# 그룹이 3개 이상일 때 사용

four_sessions = pd.read_csv(FOUR_SESSIONS_CSV)

ax = four_sessions.boxplot(by='Page', column='Time', figsize=(4, 4))
ax.set_xlabel('Page')
ax.set_ylabel('Time (in seconds)')
plt.suptitle('')
plt.title('')

plt.tight_layout()
plt.show()
````

````python
print(pd.read_csv(FOUR_SESSIONS_CSV).head())
````

````python
observed_means = four_sessions.groupby('Page').mean().values.ravel()
observed_variance = four_sessions.groupby('Page').mean().var()[0]
print('Observed means:', observed_means)
print('Observed variance:', observed_variance)

# Permutation test example with stickiness
def perm_test(df): # 분산에 대한 순열검정
    df = df.copy()
    df['Time'] = np.random.permutation(df['Time'].values)
    return df.groupby('Page').mean().var()[0]
    
print('Perm tested variance: ', perm_test(four_sessions))
````

````python
# random.seed(1)
perm_variance = [perm_test(four_sessions) for _ in range(3000)]
print('Pr(=P-value)', np.mean([var > observed_variance for var in perm_variance])) # 재표본추출한 분산이 관찰된 변화를 초과한 시간

fig, ax = plt.subplots(figsize=(5, 5))
ax.hist(perm_variance, bins=11, rwidth=0.9)
ax.axvline(x = observed_variance, color='black', lw=2)
ax.text(60, 200, 'Observed\nvariance', bbox={'facecolor':'white'})
ax.set_xlabel('Variance')
ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
````

## F-Statistic
We can compute an ANOVA table using statsmodel.

````python
model = smf.ols('Time ~ Page', data=four_sessions).fit() # ols: Ordinary Least Squares
                
aov_table = sm.stats.anova_lm(model)
print(aov_table) # df: 자유도, sum_sq: 제곱합, mean_sq: 평균제곱편차, F: F-value, PR(>F): F보다 클 확률
````

````python
res = stats.f_oneway(four_sessions[four_sessions.Page == 'Page 1'].Time, 
                     four_sessions[four_sessions.Page == 'Page 2'].Time,
                     four_sessions[four_sessions.Page == 'Page 3'].Time,
                     four_sessions[four_sessions.Page == 'Page 4'].Time)
print(f'F-Statistic: {res.statistic / 2:.4f}') # F-value: (그룹 간 차이 정도) / (불확실도)
# F-value가 높다 -> 집단 간 평균이 서로 다르다 -> 특성끼리 유사하지 않다
print(f'p-value: {res.pvalue / 2:.4f}') # 0.05보다 작으므로 통계적으로 유의미하다.
````

# Chi-Square Test
## Chi-Square Test: A Resampling Approach

````python
# 카이제곱검정: 관측된 빈도가 기대되는 빈도와 의미있게 다른지의 여부를 검정

# Table 3-4 -> 각기 다른 헤드라인에 대한 관측값
click_rate = pd.read_csv(CLICK_RATE_CSV)
clicks = click_rate.pivot(index='Click', columns='Headline', values='Rate')
print(clicks)
````

````python
# Table 3-5 -> 세 가지 헤드라인이 모두 같은 클릭률을 갖는다고 가정했을 때의 기댓값
row_average = clicks.mean(axis=1)
pd.DataFrame({
    'Headline A': row_average,
    'Headline B': row_average,
    'Headline C': row_average,
})
````

````python
# Resampling approach
box = [1] * 34 # 1: click
box.extend([0] * 2966) # 0: no-click
random.shuffle(box)

def chi2(observed, expected):
    pearson_residuals = []
    for row, expect in zip(observed, expected):
        pearson_residuals.append([(observe - expect) ** 2 / expect
                                  for observe in row])
    # return sum of squares
    return np.sum(pearson_residuals)

expected_clicks = 34 / 3
expected_noclicks = 1000 - expected_clicks
expected = [34 / 3, 1000 - 34 / 3]
chi2observed = chi2(clicks.values, expected) # 피어슨 잔차들의 제곱합

def perm_fun(box):
    sample_clicks = [sum(random.sample(box, 1000)),
                     sum(random.sample(box, 1000)),
                     sum(random.sample(box, 1000))]
    sample_noclicks = [1000 - n for n in sample_clicks]
    return chi2([sample_clicks, sample_noclicks], expected)

perm_chi2 = [perm_fun(box) for _ in range(2000)]

resampled_p_value = sum(perm_chi2 > chi2observed) / len(perm_chi2) # 재표본추출을 통해 얻은 편차의 제곱합이 얼마나 자주 관측값을 초과하는가?
print(f'Observed chi2: {chi2observed:.4f}') # 카이제곱통계량: 예측 빈도와 관측 빈도 사이의 차
print(f'Resampled p-value: {resampled_p_value:.4f}') # 재표본추출로 얻은 p값: 0.05보다 크므로 통계적으로 유의미하지 않다.
````

````python
chisq, pvalue, df, expected = stats.chi2_contingency(clicks)
print(f'Observed chi2: {chisq:.4f}')
print(f'p-value: {pvalue:.4f}') # 카이제곱분포에 대한 p값: 재표본추출로 얻은 p값과 다름 -> 카이제곱분포가 근사치이기 때문
# 카이제곱분포의 의의는 재표본 검정의 좋은 근사치를 제공하는 것
````

# Power and Sample Size
statsmodels has a number of methods for power calculation

see e.g.: https://machinelearningmastery.com/statistical-power-and-power-analysis-in-python/

````python
effect_size = sm.stats.proportion_effectsize(0.0121, 0.011) # 검정을 통해 판단할 수 있는 효과의 최소 크기(0.011->0.0121이므로 10% 증가)
analysis = sm.stats.TTestIndPower()
result = analysis.solve_power(effect_size=effect_size, 
                              alpha=0.05, power=0.8, alternative='larger') # alpha: 유의수준(5%)
print('Sample Size: %.3f' % result) # 80%의 검정력을 원할 때 116602개의 sample이 필요
````

````python
effect_size = sm.stats.proportion_effectsize(0.0165, 0.011) # 효과크기: 0.011->0.0165이므로 50% 증가
analysis = sm.stats.TTestIndPower()
result = analysis.solve_power(effect_size=effect_size, 
                              alpha=0.05, power=0.8, alternative='larger')
print('Sample Size: %.3f' % result) # 80%의 검정력을 원할 때 5488개의 sample이 필요 -> 효과크기와 필요한 sample의 수는 반비례
````
